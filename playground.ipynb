```python
import pandas as pd
```

```python
after_parse = pd.read_csv('../global_itog.csv')
after_parse.drop(columns=['Месяц'], inplace=True)
after_parse['date'] = pd.to_datetime(after_parse['date'])
after_parse.drop(columns=['url', 'Unnamed: 0'], inplace=True)
after_parse.head()
```

```python
old_df = pd.read_csv("../old_df.csv")
old_df.head()
old_df.drop(columns=['Квартал'], inplace=True)
old_df['date'] = pd.to_datetime(old_df['date'])

old_df = old_df[after_parse.columns]
# old_df.drop(columns='index', inplace=True)

old_df.head()
```

```python
fin = pd.concat([old_df, after_parse]).drop_duplicates(subset=['channel_name','ad_title','date'], keep='last')
```

```python
fin['Месяц'] = pd.Series(fin['date'].dt.month)
fin.loc[fin['Месяц'].isin([2, 3]), 'Месяц'] = 1
fin['Месяц'].unique()
```

```python
kpi_df = pd.read_excel('../data/kpi_df.xlsx')
specific_id = pd.read_excel('../data/specific_id.xlsx')
kpi_df.drop('ФО', axis=1, inplace=True)
```

```python

specific_id.columns = specific_id.columns.str.strip()
specific_id['Ссылка на канал'] = specific_id['Ссылка на канал'].str.lower()
specific_id['Ссылка на канал'] = specific_id['Ссылка на канал'].str.replace('https://t.me/', '')
specific_id['Ссылка на канал'] = specific_id['Ссылка на канал'].str.replace('\t', '')
specific_id.rename(columns={'Ссылка на канал': "channel_name", 
                   "Сетки":"Сетка"}, inplace=True)

df_that_will_be_merged_at_the_end = fin.merge(specific_id, how='left', on='channel_name')
```

```python
df_that_will_be_merged_at_the_end
```

```python
df = df_that_will_be_merged_at_the_end.merge(kpi_df, how='left', on=['Месяц', 'Регион'])
df.rename(columns={'Показы': 'KPI показов', 'Вступления таргет': 'KPI подписок'}, inplace=True)
svod1= df.groupby(['Регион', 'date', 'KPI показов', 'KPI подписок', 'Месяц'], as_index=False).agg({'Joined': 'sum', "Views": 'sum'})
svod1.sort_values(by=['Регион', 'date'], inplace=True)

svod1.reset_index(drop=True, inplace=True)
```

```python
# svod1.loc[(svod1['Регион'] =='Амурская область'), ['Регион', 'KPI подписок', 'Joined', 'Месяц']]
# svod1.loc[(svod1['Регион'] =='Кабардино-Балкарская Республика'), ['Регион', 'KPI показов', 'Views', 'Месяц']]
```

```python
views_acc = []
joined_acc = []
curr_region = svod1['Регион'][0]
curr_month = svod1['Месяц'][0]
s_joinded = 0
s_views = 0

svod2 = svod1.copy()

for index, row in svod2.iterrows():
    
    if row['Месяц'] != curr_month and curr_region == row['Регион']:
        s_joinded -= svod2.iloc[index - 1]['KPI подписок']
        s_views -= svod2.iloc[index - 1]['KPI показов']
        curr_month = row['Месяц'] 

        if s_joinded < 0:
                
            svod2.loc[(svod2['Регион'] == row['Регион']) & (svod2['Месяц'] == row['Месяц']), 'KPI подписок'] -= s_joinded
            s_joinded = 0

        if s_views < 0:
            svod2.loc[(svod2['Регион'] == row['Регион']) & (svod2['Месяц'] == row['Месяц']), 'KPI показов'] -= s_views
            s_views = 0

    
    if row['Регион'] != curr_region:
        s_joinded = 0
        s_views = 0
        curr_region = row['Регион']
        curr_month = row['Месяц'] 

    joined_acc.append(row['Joined'] + s_joinded)
    s_joinded += row['Joined']

    views_acc.append(row['Views'] + s_views)
    s_views += row['Views']


svod2['AccViews'] = views_acc
svod2['AccJoined'] = joined_acc

```

```python
# svod2.loc[(svod1['Регион'] =='Кабардино-Балкарская Республика'), ['Регион', 'KPI подписок', 'Joined', 'AccJoined', 'Месяц']]
# svod2.loc[(svod1['Регион'] =='Кабардино-Балкарская Республика'), ['Регион', 'KPI показов', 'AccViews', 'Views', 'Месяц']]
```

```python
svod2.drop(columns=['Месяц','Joined','Views'], inplace=True)
```

```python
svod2.to_excel('fin_table_n_new_kpi.xlsx', index=False)
```
